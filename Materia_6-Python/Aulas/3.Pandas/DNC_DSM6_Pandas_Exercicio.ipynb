#### 6. Qual é o BasePay médio?

```python
df["BasePay"]
```

```python
# Convert 'BasePay' to numeric, coercing errors to NaN
df['BasePay'] = pd.to_numeric(df['BasePay'], errors='coerce')

media_BasePay = df['BasePay'].mean()
print(f'A média do BasePay é: {media_BasePay}')
```

```python
df[df["BasePay"] != "NaN"]["BasePay"].astype(float).mean()
```

#### 7. Qual é o maior OvertimePay do dataset?

```python
df_TimePay = df[['Id', "EmployeeName", "OvertimePay"]]
df_TimePay
```

```python
df_TimePay["OvertimePay"] = pd.to_numeric(df_TimePay["OvertimePay"], errors="coerce")
```

```python
df_MaxTimePay = df_TimePay.sort_values(by="OvertimePay", ascending=False)
df_MaxTimePay.head(1)
```

```python
df_TimePay["OvertimePay"].max()
```

```python
df_TimePay.loc[df_TimePay["OvertimePay"].idxmax(), ["Id", "EmployeeName", "OvertimePay"]]
```

#### 8. Qual é o cargo de JOSEPH DRISCOLL?

```python
df["EmployeeName_UPPER"] = df["EmployeeName"].str.upper()
```

```python
df[df["EmployeeName_UPPER"] == "JOSEPH DRISCOLL"]
```

#### 9. Quanto ele ganha, incluindo benefícios?

```python
total_pay = df[df["EmployeeName_UPPER"] == "JOSEPH DRISCOLL"]["TotalPayBenefits"]
total_pay
```

#### 10. Quem é a pessoa que tem o maior salário base?

```python
pd.to_numeric(df["BasePay"], errors="coerce") ## ou df["BasePay"] = df["BasePay"].astype(float)

df.groupby("EmployeeName").max("BasePay").sort_values(by="BasePay", ascending=False).head(3)
```

#### 11. Quem é a pessoa mais mal paga?

```python
df.groupby("EmployeeName").max("BasePay").sort_values(by="BasePay", ascending=True).head(3)
```

#### 12. Qual a média salarial por ano?


```python
df.groupby("Year").mean("BasePay")
```

#### 13. Quantos títulos de trabalho existem?

```python
len(df["JobTitle"].unique())
```

```python
# Count the number of unique job titles (case-insensitive)
unique_JobTitles = df["JobTitle"].str.upper().nunique()
print(f"O número de títulos de trabalho únicos é: {unique_JobTitles}")
```

#### 14. Quais são os cinco principais empregos mais comuns?

```python
df.groupby("JobTitle").count().sort_values(by="Id", ascending=False).head(5)
```

```python
df["JobTitle"].value_counts().head(5)
```

```python
df["JobTitle_UPPER"] = df["JobTitle"].str.upper()
df.groupby("JobTitle_UPPER").count().sort_values(by="Id", ascending=False).head(5)
```

```python
count_JobTitles = df["JobTitle"].str.upper()
count_JobTitles.value_counts().head(5)
```

#### 15. Quantas pessoas têm a palavra chefe em seu cargo? (complicado)

```python
def chefe (valor):
  if "chief" in valor.lower():
    return True
  else:
    return False
```

```python
df["JobTitle"].apply(chefe)
```

```python
df[df["JobTitle"].apply(chefe)]
```

```python
df[df["JobTitle"].apply(chefe)].count()
```
