#### 6. Qual é o BasePay médio?

```python
df["BasePay"]
```

```python
# Convert 'BasePay' to numeric, coercing errors to NaN
df['BasePay'] = pd.to_numeric(df['BasePay'], errors='coerce')

media_BasePay = df['BasePay'].mean()
print(f'A média do BasePay é: {media_BasePay}')
```

```python
df[df["BasePay"] != "NaN"]["BasePay"].astype(float).mean()
```

#### 7. Qual é o maior OvertimePay do dataset?

```python
df_TimePay = df[['Id', "EmployeeName", "OvertimePay"]]
df_TimePay
```

```python
df_TimePay["OvertimePay"] = pd.to_numeric(df_TimePay["OvertimePay"], errors="coerce")
```

```python
df_MaxTimePay = df_TimePay.sort_values(by="OvertimePay", ascending=False)
df_MaxTimePay.head(1)
```

```python
df_TimePay["OvertimePay"].max()
```

```python
df_TimePay.loc[df_TimePay["OvertimePay"].idxmax(), ["Id", "EmployeeName", "OvertimePay"]]
```
