### 1. üß† Ficha de Revis√£o R√°pida | (O Essencial)

<br>

> - **Vari√°vel Categ√≥rica:** frequ√™ncia absoluta (`value_counts()`) e relativa (`value_counts(normalize=True)`).  
> - **Vari√°vel Num√©rica:** histogramas (`hist()`) e contagem por bin (`np.histogram`).  
> - **Propor√ß√£o:** converte contagem em percentual para compara√ß√µes.  
> - **Conting√™ncia:** tabelas cruzadas (`pd.crosstab`) para dupla frequ√™ncia.  

<br>


### 2. üî¨ Mergulho Profundo | (Os Detalhes)

<br>

#### **üéØ O Conceito Central**  
Calcular frequ√™ncia transforma dados brutos em distribui√ß√£o de ocorr√™ncias:  
1. Para **categ√≥ricas**, contamos quantas vezes cada categoria aparece.  
2. Para **num√©ricas**, dividimos o dom√≠nio em *bins* (intervalos) e contamos quantos valores caem em cada um.  
3. **Frequ√™ncia relativa** normaliza para propor√ß√µes ou percentuais, essencial em amostras de tamanhos distintos.  
4. Tabelas de **conting√™ncia** cruzam duas vari√°veis categ√≥ricas, exibindo frequ√™ncia conjunta.

<br>

#### **üîó Analogia de Data Science**  
√â como contar quantos pacotes de cada cor chegam por dia (categ√≥ricas) e agrupar as idades dos clientes em faixas de 0‚Äì10, 10‚Äì20 anos etc. (num√©ricas), depois transformar em porcentagem para relat√≥rios.

<br>

#### **N√≠vel DEUS (2/3): Fun√ß√£o Distribui√ß√£o Emp√≠rica (ECDF)**


````
import pandas as pd
import numpy as np
import statsmodels.api as sm

# ‚ÄúLeia dados.‚Äù
df = pd.read_csv('/content/ifood-restaurants-november-2020.csv')

# ‚ÄúCalcule ECDF para delivery_time.‚Äù
ecdf = sm.distributions.ECDF(df['delivery_time'])
x = np.linspace(df['delivery_time'].min(), df['delivery_time'].max(), 100)
y = ecdf(x)

# ‚ÄúMonte DataFrame e mostre.‚Äù
ecdf_df = pd.DataFrame({'delivery_time': x, 'ecdf': y})
print(ecdf_df.head())
````

### **Resposta R√°pida**

Use **frequ√™ncia relativa** quando quiser **comparar distribui√ß√µes entre grupos de tamanhos diferentes** ‚Äî ela mostra **propor√ß√µes**, n√£o contagens brutas.

---

### **Analogia do Dia**

Imagine duas salas de aula:

* Sala A tem 10 alunos, e 5 tiraram nota A ‚Üí 50%
* Sala B tem 100 alunos, e 30 tiraram nota A ‚Üí 30%

Mesmo que **Sala B tenha mais alunos com nota A (frequ√™ncia absoluta)**, **Sala A tem maior propor√ß√£o de excel√™ncia (frequ√™ncia relativa)**.
