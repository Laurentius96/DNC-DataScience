### Exerc√≠cio 3 - An√°lise de Frequ√™ncia para Dados Qualitativos

<br>
Uma empresa coletou feedbacks de clientes sobre um novo produto. Os resultados  foram:

<br>

```Python
feedbacks = ["Bom", "Ruim", "Excelente", "Bom", "Regular", "Bom", "Excelente", "Ruim", "Bom", "Regular"]
```

<br>

üìå Passos:
1. Construa uma tabela de frequ√™ncia para os feedbacks.
2. Qual √© a categoria mais frequente?
3. Represente os dados em um gr√°fico de barras.

````python
feedbacks = ["Bom", "Ruim", "Excelente", "Bom", "Regular", "Bom", "Excelente", "Ruim", "Bom", "Regular"]
````

````python
df = pd.DataFrame(feedbacks, columns=['feedbacks'])
df
````

````python
## 1. Construa uma tabela de frequ√™ncia para os feedbacks.
tabela = df.groupby('feedbacks').size().reset_index(name='counts')
tabela
````

````python
## 2. Qual √© a categoria mais frequente?

'''
Bom.
'''
````

````python
## 3. Represente os dados em um gr√°fico de barras.

plt.figure(figsize=(10, 6))
plot = sns.barplot(x='feedbacks', y='counts',  data=tabela)
for i in plot.patches:
    plot.annotate(f'{i.get_height()}', (i.get_x() + i.get_width() / 2, i.get_height()), ha='center', va='center', size=10, xytext=(0, 5), textcoords='offset points')
plt.xlabel('Feedbacks')
plt.ylabel('Frequ√™ncia')
plt.title('Tabela de Frequ√™ncia dos Feedbacks')
plt.show()
````

## 5. Compara√ß√£o entre An√°lise Quantitativa e Qualitativa

<br>

Um analista de dados quer verificar a rela√ß√£o entre o n√∫mero de horas de estudo e a nota final em um exame.

<br>

```
horas_estudo = [2, 4, 6, 8, 10, 12, 14]

notas = [50, 55, 65, 70, 80, 85, 90]

```

<br>

üìå Passos:
1. Calcule a correla√ß√£o entre horas de estudo e nota final.
2. A rela√ß√£o entre essas vari√°veis √© positiva 2. ou negativa?
3. Qual conclus√£o pode ser tirada a partir desse resultado?




````python
horas_estudo = [2, 4, 6, 8, 10, 12, 14]
notas = [50, 55, 65, 70, 80, 85, 90]
````

````python
## 1. Calcule a correla√ß√£o entre horas de estudo e nota final.

correlation = np.corrcoef(horas_estudo, notas)[0, 1]
print(f"Correla√ß√£o: {correlation}")


````

````python
## 2. A rela√ß√£o entre essas vari√°veis √© positiva 2. ou negativa?

'''
Positiva forte
'''
````

````python
## 3. Qual conclus√£o pode ser tirada a partir desse resultado?

'''
Existe uma rela√ß√£o direta, indicando que estudar mais melhora o desempenho no exame.
'''
````
